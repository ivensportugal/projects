```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn import tree

from sklearn.metrics import accuracy_score

df = pd.read_csv('grade2.csv', sep='\t')
X = df[['grade1','grade2','grade3','grade4','grade5']]
y = df['pass']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)

m = tree.DecisionTreeClassifier()
m = m.fit(X_train,y_train)

p = m.predict(X_test)
a = accuracy_score(y_test,p)

print('accuracy: ', a)

d = pd.DataFrame(np.array([0,0,0,0,10]).reshape(1,5), columns=['grade1','grade2','grade3','grade4','grade5'])
p = m.predict(d)
print('prediction: ',p)

plt.figure(figsize=(20,20))  # Adjust the figure size as needed
tree.plot_tree(m,feature_names=['grade1','grade2','grade3','grade4','grade5'],class_names=['fail','pass'])
plt.show()
```
