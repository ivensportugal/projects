```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.cluster import KMeans

df = pd.read_csv('coord.csv',sep='\t')

X = df[['x','y']]
y = df['q']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.4)

m = KMeans(n_clusters=4)
m = m.fit(X_train)
p = m.predict(X_test)
a = accuracy_score(p,y_test)
print('accuracy: ', a)

d = pd.DataFrame(np.array([3,2]).reshape(1,2), columns=df.columns[:2])
p = m.predict(d)
print('prediction: ', p[0])

# Plot data points
plt.scatter(X_train['x'], X_train['y'], c=m.labels_, cmap='viridis', s=50, alpha=0.5)

# Plot centroids
plt.scatter(m.cluster_centers_[:, 0], m.cluster_centers_[:, 1], c='red', marker='x', s=200, label='Centroids')

plt.title('KMeans Clustering')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
